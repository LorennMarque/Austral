```python
import matplotlib.pyplot as plt
import numpy as np
import csv

def ej_1():
    banks_dict = {}
    with open('bancos.csv', 'r') as bankdata:
        data = csv.reader(bankdata)
        next(data)  # Skip header row if there is one
        for row in data:
            bank_name = row[0]
            date = row[1]
            value = float(row[5])  # Assuming the values are numerical
            if bank_name not in banks_dict:
                banks_dict[bank_name] = {'dates': [], 'values': []}
            banks_dict[bank_name]['dates'].append(date)
            banks_dict[bank_name]['values'].append(value)
    return banks_dict

def plot_banks_data(banks_dict):
    num_banks = len(banks_dict)
    fig, axs = plt.subplots(num_banks, 1, figsize=(10, 5 * num_banks))

    if num_banks == 1:
        axs = [axs]  # Ensure axs is iterable if there's only one subplot

    for ax, (bank_name, data) in zip(axs, banks_dict.items()):
        ax.plot(data['dates'], data['values'], marker='o', linestyle='-', label=bank_name)
        ax.set_title(bank_name)
        ax.set_xlabel('Date')
        ax.set_ylabel('Value')
        ax.legend()
        ax.grid(True)

    plt.tight_layout()
    plt.show()

# Load and parse the data
banks_data = ej_1()

# Plot the data
plot_banks_data(banks_data)
```
