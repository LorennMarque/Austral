El proceso de clculo de la correlacin entre las columnas numricas de un DataFrame y la generacin de una nueva matriz de correlacin es una tarea fundamental en el anlisis de datos, particularmente til para entender cmo las variables numricas en un conjunto de datos se relacionan entre s. Vamos a desglosar este proceso en ms detalle, enfocndonos en el uso del coeficiente de correlacin de Pearson, que es el mtodo predeterminado en muchos entornos de anlisis de datos, incluido el mtodo `.corr()` en pandas.

### Clculo de la Matriz de Correlacin

1. **Seleccin de Columnas Numricas:** Primero, el mtodo `.corr()` identifica y selecciona todas las columnas numricas presentes en el DataFrame. Solo se consideran estas columnas porque la correlacin, como medida estadstica, se aplica a variables cuantitativas.

2. **Aplicacin del Coeficiente de Correlacin de Pearson:** Para cada par de columnas numricas, se calcula el coeficiente de correlacin de Pearson. Este coeficiente es un nmero entre -1 y 1 que mide el grado de relacin lineal entre dos variables. Se calcula como la covarianza de las dos variables dividida por el producto de sus desviaciones estndar. Matemticamente, se expresa como:

   \[ r = \frac{n(\sum xy) - (\sum x)(\sum y)}{\sqrt{[n\sum x^2 - (\sum x)^2][n\sum y^2 - (\sum y)^2]}} \]

   donde:
   - \( r \) es el coeficiente de correlacin de Pearson.
   - \( n \) es el nmero de observaciones.
   - \( x \) e \( y \) son las variables que se estn comparando.
   - \( \sum \) denota la suma sobre todas las observaciones.

3. **Interpretacin del Coeficiente:**
   - **1 o -1:** Una correlacin de 1 indica una relacin lineal perfecta positiva, donde los valores de ambas variables se mueven en la misma direccin. Una correlacin de -1 indica una relacin lineal perfecta negativa, donde los valores de una variable se mueven en direccin opuesta a la otra.
   - **0:** Una correlacin de 0 indica que no hay relacin lineal entre las variables.
   - **Valores intermedios:** Valores cercanos a 1 o -1 indican una fuerte relacin lineal, mientras que valores cercanos a 0 indican una relacin lineal dbil.

### Especificacin de Otros Mtodos

Aunque el coeficiente de Pearson es el predeterminado para `.corr()`, se pueden especificar otros mtodos para calcular la correlacin, cada uno con sus propias aplicaciones y suposiciones. Entre estos, los ms comunes son:
- **Spearman:** Usado para datos ordinales o para evaluar relaciones monotnicas no lineales.
- **Kendall:** Similar a Spearman, pero ms robusto y con una interpretacin ligeramente diferente en trminos de la concordancia entre pares de observaciones.

Al finalizar el clculo para cada par de columnas, se genera una matriz de correlacin donde cada elemento \( i, j \) representa la correlacin entre la columna \( i \) y la columna \( j \), proporcionando una visin completa de cmo todas las variables numricas en el DataFrame se relacionan entre s.

Tendencias por desenlace.

### 4.2.1. Organizar el dataset

````python
def timeseries_data(data_median_hour, variable, filldata = 1):
    data4clustering = data_median_hour.pivot(columns='hour', values=variable)
    if filldata == 1:
        # first forward fill
        data4clustering = data4clustering.fillna(method='ffill', axis=1)
        # next backward fill
        data4clustering = data4clustering.fillna(method='bfill', axis=1)
        data4clustering = data4clustering.dropna(axis=0, how='all')
    return data4clustering
````

### 4.2.2. Grficos de valores medios por desenlace.

# 5. Feature engineering
## 5.1. Construccin de features (variables)

Se crean variables para obtener la mayor cantidad de informacin posible a partir de los datos.

Para el caso de las series temporales, se podran construir las siguientes features: mximo, mnimo, media, desvo estndar.

OJO! Cuanta mayor cantidad de feature agrego, se agranda el dataset!

````python
def feat_transf(data):
    data_max = data.groupby(['icustay'])[variables].max()
    data_max.columns = ['max ' + str(col) for col in data_max.columns]

    data_min = data.groupby(['icustay'])[variables].min()
    data_min.columns = ['min ' + str(col) for col in data_min.columns]

    data_sd = data.groupby(['icustay'])[variables].std()
    data_sd.columns = ['sd ' + str(col) for col in data_sd.columns]

    data_mean = data.groupby(['icustay'])[variables].mean()
    data_mean.columns = ['mean ' + str(col) for col in data_mean.columns]

    data_agg = pd.concat([data_min,data_max,data_sd,data_mean],axis=1)

    return data_agg

data_transf = feat_transf(data_median_hour)
print('Extracted features: ')
display.display(list(data_transf.columns))
print('')
print('Number of ICU stays: ' + str(data_transf.shape[0]))
print('Number of features: ' + str(data_transf.shape[1]))
display.display(data_transf.head())
````
