El proceso de cálculo de la correlación entre las columnas numéricas de un DataFrame y la generación de una nueva matriz de correlación es una tarea fundamental en el análisis de datos, particularmente útil para entender cómo las variables numéricas en un conjunto de datos se relacionan entre sí. Vamos a desglosar este proceso en más detalle, enfocándonos en el uso del coeficiente de correlación de Pearson, que es el método predeterminado en muchos entornos de análisis de datos, incluido el método `.corr()` en pandas.

### Cálculo de la Matriz de Correlación

1. **Selección de Columnas Numéricas:** Primero, el método `.corr()` identifica y selecciona todas las columnas numéricas presentes en el DataFrame. Solo se consideran estas columnas porque la correlación, como medida estadística, se aplica a variables cuantitativas.

2. **Aplicación del Coeficiente de Correlación de Pearson:** Para cada par de columnas numéricas, se calcula el coeficiente de correlación de Pearson. Este coeficiente es un número entre -1 y 1 que mide el grado de relación lineal entre dos variables. Se calcula como la covarianza de las dos variables dividida por el producto de sus desviaciones estándar. Matemáticamente, se expresa como:

   \[ r = \frac{n(\sum xy) - (\sum x)(\sum y)}{\sqrt{[n\sum x^2 - (\sum x)^2][n\sum y^2 - (\sum y)^2]}} \]

   donde:
   - \( r \) es el coeficiente de correlación de Pearson.
   - \( n \) es el número de observaciones.
   - \( x \) e \( y \) son las variables que se están comparando.
   - \( \sum \) denota la suma sobre todas las observaciones.

3. **Interpretación del Coeficiente:**
   - **1 o -1:** Una correlación de 1 indica una relación lineal perfecta positiva, donde los valores de ambas variables se mueven en la misma dirección. Una correlación de -1 indica una relación lineal perfecta negativa, donde los valores de una variable se mueven en dirección opuesta a la otra.
   - **0:** Una correlación de 0 indica que no hay relación lineal entre las variables.
   - **Valores intermedios:** Valores cercanos a 1 o -1 indican una fuerte relación lineal, mientras que valores cercanos a 0 indican una relación lineal débil.

### Especificación de Otros Métodos

Aunque el coeficiente de Pearson es el predeterminado para `.corr()`, se pueden especificar otros métodos para calcular la correlación, cada uno con sus propias aplicaciones y suposiciones. Entre estos, los más comunes son:
- **Spearman:** Usado para datos ordinales o para evaluar relaciones monotónicas no lineales.
- **Kendall:** Similar a Spearman, pero más robusto y con una interpretación ligeramente diferente en términos de la concordancia entre pares de observaciones.

Al finalizar el cálculo para cada par de columnas, se genera una matriz de correlación donde cada elemento \( i, j \) representa la correlación entre la columna \( i \) y la columna \( j \), proporcionando una visión completa de cómo todas las variables numéricas en el DataFrame se relacionan entre sí.

## 4.2. Ploteo de datos temporales

Tendencias por desenlace.

### 4.2.1. Organizar el dataset

````python
def timeseries_data(data_median_hour, variable, filldata = 1):
    data4clustering = data_median_hour.pivot(columns='hour', values=variable)
    if filldata == 1:
        # first forward fill
        data4clustering = data4clustering.fillna(method='ffill', axis=1)
        # next backward fill
        data4clustering = data4clustering.fillna(method='bfill', axis=1)
        data4clustering = data4clustering.dropna(axis=0, how='all')
    return data4clustering
````

### 4.2.2. Gráficos de valores medios por desenlace.

# 5. Feature engineering
## 5.1. Construcción de features (variables)

Se crean variables para obtener la mayor cantidad de información posible a partir de los datos.

Para el caso de las series temporales, se podrían construir las siguientes features: máximo, mínimo, media, desvío estándar.

OJO! Cuanta mayor cantidad de feature agrego, se agranda el dataset!

````python
def feat_transf(data):
    data_max = data.groupby(['icustay'])[variables].max()
    data_max.columns = ['max ' + str(col) for col in data_max.columns]

    data_min = data.groupby(['icustay'])[variables].min()
    data_min.columns = ['min ' + str(col) for col in data_min.columns]

    data_sd = data.groupby(['icustay'])[variables].std()
    data_sd.columns = ['sd ' + str(col) for col in data_sd.columns]

    data_mean = data.groupby(['icustay'])[variables].mean()
    data_mean.columns = ['mean ' + str(col) for col in data_mean.columns]

    data_agg = pd.concat([data_min,data_max,data_sd,data_mean],axis=1)

    return data_agg

data_transf = feat_transf(data_median_hour)
print('Extracted features: ')
display.display(list(data_transf.columns))
print('')
print('Number of ICU stays: ' + str(data_transf.shape[0]))
print('Number of features: ' + str(data_transf.shape[1]))
display.display(data_transf.head())
````
